# Example: BSpline Package

```python
import numpy as np
import matplotlib.pyplot as plt
import sys
sys.path.append('../bspline/')
from bspline import bspline
from lbspline import lbspline
```

## Create Test Data

```python
knots = np.linspace(1.0, 2.0, 5)
degree = 3
```

### Test `bspline`

```python
a = np.repeat(knots[0], 201)
x = np.linspace(knots[0], knots[-1], 201)
bs = bspline(knots, degree)

X = bs.designMat(x)
DX = bs.designDMat(x, 1)
IX = bs.designIMat(a, x, 1)
```

```python
fig, ax = plt.subplots(3, 1, figsize=(8, 12))
ax[0].plot(x, X)
ax[0].set_title('spline bases')

ax[1].plot(x, DX)
ax[1].set_title('spline basis derivatives')

ax[2].plot(x, IX)
ax[2].set_title('spline basis integrals')
```

### Test `lbspline`

```python
a = np.repeat(knots[0], 201)
x = np.linspace(knots[0], knots[-1], 201)
lbs = lbspline(knots, degree, l_linear=True, r_linear=True)

X = lbs.designMat(x)
DX = lbs.designDMat(x, 1)
IX = lbs.designIMat(a, x, 1)
```

```python
fig, ax = plt.subplots(3, 1, figsize=(8, 12))
ax[0].plot(x, X)
ax[0].set_title('spline bases')

ax[1].plot(x, DX)
ax[1].set_title('spline basis derivatives')

ax[2].plot(x, IX)
ax[2].set_title('spline basis integrals')
```
